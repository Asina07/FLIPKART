```python
#importing all packages
import time
import requests
from IPython.display import clear_output
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Website link
base_url= 'https://www.flipkart.com/mobiles/mi~brand/pr?sid=tyy,4io&otracker=nmenu_sub_Electronics_0_Mi'
```

```python
pname=[]  #product name
pprc=[]   #product price
sr=[]     #star rating
rr=[]     #rating and review
features=[]    #description  

for i in range(1,3):
  time.sleep(3)
#Reqest from server 
url=base_url+str(i)
req=requests.get(url)
soup=BeautifulSoup(req.text,'html')

```

```python
#Name of the Mobile
pro_name = soup.findAll('div',attrs={"class":"_4rR01T"})
for j in pro_name:
  pname.append(j.text)

#price of mobile
p_prc = soup.findAll('div' , attrs={"class" : "_3tbKJL"})  
for j in p_prc:
  pprc.append(j.text)

#Star Rating
star_rate = soup.findAll('div', attrs ={"class" : "gUuXy-"}) 
for j in star_rate:
  sr.append(j.text)  

#rate & reviews
rate_rew=soup.findAll('span',attrs={'class':'_2_R_DZ'})      
for j in rate_rew:
  r=j.text.split('&')
  rt.append(r[0].split(' ')[0])
  rr.append(r[1].split(' ')[0].strip())

#features
a=soup.findAll('ul',attrs={'class':'_1xgFaf'})                      
for j in a:
  a=j.findAll('li')
  features.append([j.text for j in a])
  
 

```

```python
len(pname),len(pprc),len(sr),len(rr),len(features)
```

```python
#Converting to dataframe
data = pd.DataFrame ({'Name' : pname, 'price' : pprc, 'raterew' : rr   })
data
```

```python
data.to_csv('flipkart_data.csv',index=False)
```
